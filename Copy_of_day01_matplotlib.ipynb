# Visualizing Data with Matplotlib

Visualization is one of the most important steps in data analysis — a good plot can reveal patterns that are invisible in raw numbers.

We'll use **Matplotlib**, Python's most widely-used plotting library. You've already used `plt.imshow()` to display images in the NumPy notebook — now we'll learn the four plot types you'll use most as a scientist:

1. **Line plots** — trends over time
2. **Scatter plots** — relationships between two variables
3. **Histograms** — the distribution of a single variable
4. **Bar charts** — comparing values across categories

See the [Matplotlib documentation](https://matplotlib.org/stable/) for a full reference.

## Setup: Loading our data

We'll use the [Gapminder](https://www.gapminder.org/) dataset from the pandas notebook.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def load_gapminder():
    url = "https://raw.githubusercontent.com/plotly/datasets/master/gapminder_unfiltered.csv"
    try:
        return pd.read_csv(url)
    except Exception as e:
        raise RuntimeError(
            "Could not download the dataset. Check your internet connection "
            "and re-run this cell. (If this persists, ask an instructor.)"
        ) from e

gapminder = load_gapminder()

# We'll focus on the most recent year for most plots
data_2007 = gapminder[gapminder["year"] == 2007]

gapminder.head()
```

> **Reminder:** The cells above must be run before anything else in this notebook works. If you see a `NameError`, come back here and run them first.

---
## 1. Line Plots

Line plots are the go-to for showing how something changes over time.

```python
# How has life expectancy changed over time in the US?
usa = gapminder[gapminder["country"] == "United States"]

plt.plot(usa["year"], usa["lifeExp"], label = "USA")
plt.plot(usa["year"], usa["lifeExp"]+1, label = "USA + 1")
plt.xlabel("Year")
plt.ylabel("Life Expectancy (years)")
plt.title("Life Expectancy in the United States")
plt.legend()
```

**Always label your axes and add a title.** A plot without labels is hard for others (and your future self) to interpret.

You can compare multiple lines by calling `plt.plot()` several times. Use `label=` and `plt.legend()` to tell them apart.

---
## 2. Scatter Plots

Scatter plots reveal **relationships** between two numerical variables. This is the plot you'll reach for most often when exploring data.

**🤔 Guess first:** Each dot in the next plot is one country in 2007. Before running it, predict: as GDP per capita goes up, does life expectancy go **up**, go **down**, or show **no pattern**?


```python
# Does money buy longevity? GDP per capita vs life expectancy, one dot per country
plt.scatter(data_2007["gdpPercap"], data_2007["lifeExp"], alpha=0.6)
plt.xlabel("GDP per Capita (USD)")
plt.ylabel("Life Expectancy (years)")
plt.title("GDP vs Life Expectancy (2007)")
```

**Always label your axes and add a title.** A plot without labels is hard for others (and your future self) to interpret.

There seems to be a relationship, but it's not linear — it curves. When data spans a huge range (GDP goes from ~$300 to ~$50,000), a **log scale** can help.

**🤔 Guess first:** The next cell re-draws the same scatter with `plt.xscale("log")`. Before running it, predict how the shape of the point cloud will change — will the curve straighten out, or bend more?


```python
plt.scatter(data_2007["gdpPercap"], data_2007["lifeExp"], alpha=0.6)
plt.xscale("log")  # Logarithmic scale — spaces values by powers of 10
plt.xlabel("GDP per Capita (USD, log scale)")
plt.ylabel("Life Expectancy (years)")
plt.title("GDP vs Life Expectancy (2007)")
```

---
## 3. Histograms

Histograms show the **distribution** of a single variable — how frequently different values occur. This is essential for understanding your data before doing any modeling.

```python
plt.hist(data_2007["lifeExp"], bins=20, edgecolor="white", color="teal")
plt.xlabel("Life Expectancy (years)")
plt.ylabel("Number of Countries")
plt.title("Distribution of Life Expectancy (2007)")
```

---
## 4. Bar Charts

Bar charts compare values across **categories**. They pair naturally with pandas `groupby` results.

**🤔 Guess first:** Two predictions before running the next cell: **(1)** which continent do you think has the *highest* average life expectancy in 2007? **(2)** The code sorts with `.sort_values()` (ascending) — so will that continent's bar appear *first* or *last*?


```python
# Average life expectancy by continent in 2007
continent_avg = data_2007.groupby("continent")["lifeExp"].mean().sort_values()

plt.bar(continent_avg.index, continent_avg.values, color="steelblue")
plt.ylabel("Average Life Expectancy (years)")
plt.title("Average Life Expectancy by Continent (2007)")
plt.xticks(rotation=45)
plt.tight_layout()
```

---
## 5. Saving Figures

Once you've made a figure you're happy with, save it to a file with `plt.savefig()`.

**🤔 Guess first:** Before running the next cell, predict: where will the file `gdp_vs_life_expectancy.png` end up? After running it, check the file browser (the folder icon in the sidebar) to see if you were right.


```python
plt.scatter(data_2007["gdpPercap"], data_2007["lifeExp"], alpha=0.6)
plt.xscale("log")
plt.xlabel("GDP per Capita (USD, log scale)")
plt.ylabel("Life Expectancy (years)")
plt.title("GDP vs Life Expectancy (2007)")

plt.savefig("gdp_vs_life_expectancy.png", dpi=150)
print("Saved!")
```

You can save as `.png`, `.pdf`, `.svg`, etc. The `dpi` parameter controls resolution.

## Key Points

- Import matplotlib with `import matplotlib.pyplot as plt`.
- **Line plots** (`plt.plot()`) show trends over time. Use `label=` and `plt.legend()` to compare multiple lines.
- **Scatter plots** (`plt.scatter()`) show relationships between two variables.
- **Histograms** (`plt.hist()`) show the distribution of a single variable.
- **Bar charts** (`.plot(kind="bar")`) compare values across categories.
- Always label axes (`plt.xlabel()`, `plt.ylabel()`) and add a title (`plt.title()`).
- Use `plt.xscale("log")` when data spans a large range.
- Save figures with `plt.savefig("filename.png")`.
